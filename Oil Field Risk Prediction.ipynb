**Project Brief: Optimal Drilling Location Determination for an Oil Company**

**Background and Objective**

An established oil company is planning to embark on a new drilling venture. The critical decision involves identifying the most profitable drilling location. We have data on oil samples from three different regions, each with 10,000 fields. These fields have been evaluated for oil quality and reserve volume. The main goal is to develop a machine learning model to determine the region that offers the highest profitability from oil extraction. The project also aims to analyze potential profits and risks using the Bootstrap technique.

**Task Overview**

The process for selecting the optimal drilling location includes several key steps:

1. **Regional Analysis:**
   - Search for potential deposits in the chosen region and assess characteristic values for each field.

2. **Model Development:**
   - Construct a machine learning model to estimate the volume of reserves in each field.

3. **Field Selection:**
   - Use the model to identify fields with the highest estimated reserves, considering the company's budget and the cost of developing each well.

4. **Profit Calculation:**
   - Calculate the overall profit based on the total profits from the selected fields.
   
**Deliverables**

The project will result in the following deliverables:
- A machine learning model for reserve volume estimation.
- A detailed analysis report on predicted profits and risk assessment using Bootstrap methods.
- Strategic recommendations for the most viable region for drilling based on model predictions and profit-risk analysis.

## Model training and testing

```python
warnings.filterwarnings("ignore")

# We split the data into training and validation samples in a ratio of 75:25

predictions_list = []
target_valid_list = []

for data, count in zip(data_list, range(3)):
    print(f'results by region {count}, data_{count}')
    target = data['product']
    features = data.drop(['id','product'], axis=1)

    features_train, features_valid, target_train, target_valid = train_test_split(
        features,
        target,
        test_size=0.25,
        random_state=12345)

# Scaling the features

    scaler = StandardScaler()
    scaler.fit(features_train)
    features_train = scaler.transform(features_train)
    features_valid = scaler.transform(features_valid)

# We train the model and make predictions on the validation set

    model = LinearRegression()
    
    hyperparameters = {
        'fit_intercept': [True, False],
        'copy_X': [True, False]
    }

    grid_search = GridSearchCV(model, param_grid=hyperparameters, cv=5, scoring='neg_mean_squared_error')
    grid_search.fit(features_train, target_train)
    
    print('Best hyperparameters: ', grid_search.best_params_)
# print('Best score: ', grid_search.best_score_)
    
    model = LinearRegression(**grid_search.best_params_)
    model.fit(features_train, target_train)
    predicted_valid = model.predict(features_valid)
    rse = mean_squared_error(target_valid, predicted_valid, squared=False)
    
# We save predictions and correct answers on the validation set
    
    predictions_list.append(predicted_valid)
    target_valid_list.append(target_valid)
    
# Print on the screen the average inventory of predicted raw materials and the RMSE model

    print('Average stock of predicted raw materials', predicted_valid.mean())
    print('RMSE =', rse)
    print()
```

**Conclusions on model training and testing:**
- to predict the stock of raw materials, the LinearRegression() model was used with the selection of hyperparameters 'fit_intercept', 'normalize', 'copy_X'
- the model was trained best on the second sample (data_1), RMSE = 0.89
- the smallest reserve of predicted raw materials was also obtained in the second sample (data_1)

## Preparation for profit calculation

**Key values ​​for calculations are stored in separate variables**

```python
BUDGET= 10000000000
BARREL_TH= 450000
WELL_COUNT= 200
```

**We will calculate a sufficient volume of raw materials for the break-even development of a new well**

```python
one_well_budget = BUDGET / WELL_COUNT
min_item_count = one_well_budget / BARREL_TH
print(min_item_count)
```

**Let’s compare the received volume of raw materials with the average stock in each region**

```python
for target_valid, count in zip(target_valid_list, range(3)):
    print(f'results by region {count}, data_{count}')
    print(f'The difference between the sufficient volume of raw materials and the average is: {round(target_valid.mean() - min_item_count, 3)} units')
    print()
```

**The sufficient volume of raw materials for the well turned out to be 111.11, the difference in all three samples between the average volume and the sufficient volume turned out to be negative. The biggest difference of -42.388 units was obtained in the second sample (data_1)**

**Function for calculating profit for selected wells and model predictions**

```python
def profit_(target, predictions, count):
    pred_sorted = predictions.sort_values(ascending=False)
    selected = target[pred_sorted.index][:count]
    summa = selected.sum()
    return summa * BARREL_TH - BUDGET
```

## Calculation of profits and risks

```python
state = np.random.RandomState(12345)
bootstrap_points = 500
selected_points = 200

# Loop through all three deposits

for target, predictions, count in zip(target_valid_list, predictions_list, range(3)):
    target = target.reset_index(drop=True)
    predictions = pd.Series(predictions)
    print(f'results by region {count}, data_{count}')
    
# Create variables

    values = []
    risks = []

# In the cycle we bootstrap 500 values ​​from each field 1000 times
    
    for i in range(1000):
        target_subsample = target.sample(n=bootstrap_points, replace=True, random_state=state)
        pred_subsample = predictions[target_subsample.index]
        
# We use the profit calculation function to select the 200 most profitable mines in terms of production

        well_profit = profit_(target_subsample, pred_subsample, selected_points)
        values.append(well_profit)

# We check how much negative profit was received as a result of the selected mines

# for v in values:
# if v < 0:
# risks.append(v)
    
# Calculate the 95% confidence interval and the percentage of risk of loss

    values = pd.Series(values)
    confidence_interval = st.t.interval(0.95, len(values)-1, values.mean(), values.sem())
# risk = len(risks) / 1000 * 100
    risk = (values < 0).mean() * 100

# Display data for each region


    mean = values.mean()
    print('max quantile:', values.quantile(0.975))
    print('min quantile:', values.quantile(0.025))
    print('Average revenue:', mean)
    print('95% confidence interval of the mean:', confidence_interval)
    print(f'Risk of loss: {round(risk, 2)}%')
    print()
```

**Conclusions on calculating profits and risks:**
- Based on the results of the analysis, the second region (data_1) was selected. After assessing the risks, it turned out to be the only one on the list with a probability of losses equal to 0.3% (less than 2.5%), as well as with the highest average revenue of 536 million.
- The remaining regions exceeded the figure by 2.5%, the first and second regions have a risk of losses of 7.1% and 11.8%, respectively, their average profitability is in the region of 409 and 339 million.
- 95% confidence interval showing the spread of the average in the selected region is in the region of 523-550 million.
