**Project Overview**

**Context**

Working at "Streamchik", an online game retailer, we aim to identify key factors that determine the success of computer games globally. This involves analyzing historical sales data, user and expert reviews, game genres, and platforms like Xbox and PlayStation.

**Data Timeline**

The available data covers up to 2016. We are in December 2016, planning for the year 2017. The focus is on understanding data handling principles for sales prediction, applicable for any future year.

**Key Element**

The dataset includes ESRB ratings, which assess game content and categorize it by age suitability.

**Project Objective**

To uncover patterns and trends that define the success of computer games. This will guide decisions for future product selection and advertising strategies.

**Data Description**

- `Name`: Game title.
- `Platform`: Gaming platform (e.g., Xbox, PlayStation).
- `Year_of_Release`: Year of game release.
- `Genre`: Game genre.
- `NA_sales`: Sales in North America (in millions).
- `EU_sales`: Sales in Europe (in millions).
- `JP_sales`: Sales in Japan (in millions).
- `Other_sales`: Sales in other regions (in millions).
- `Critic_Score`: Critics' score (max 100).
- `User_Score`: Users' score (max 10).
- `Rating`: ESRB rating.

**Note:** Data for 2016 may be incomplete.

### Hypotheses.

```python
alpha = 0.05
```

**H0 = Average user ratings for Xbox One and PC platforms are the same.**

**H_a = Average user ratings for Xbox One and PC platforms are different.**

```python
# from scipy import stats as st
```

```python
user_score_xbox = data_from_2013[(data_from_2013['platform'] == 'XOne') & (data_from_2013['user_score'].notna())]['user_score']
user_score_pc = data_from_2013[(data_from_2013['platform'] == 'PC') & (data_from_2013['user_score'].notna())]['user_score']

results = st.ttest_ind(
    user_score_xbox, 
    user_score_pc,
    equal_var=False)

print('p-value:', results.pvalue)

if results.pvalue < alpha:
    print("Reject the null hypothesis")
else:
    print("Failed to reject the null hypothesis")
```

**H0 = Average user ratings for the Action and Sports genres are different.**

**H_a = Average user ratings for the Action and Sports genres are the same.**

```python
user_score_action = data_from_2013[(data_from_2013['genre'] == 'Action') & (data_from_2013['user_score'].notna())]['user_score']
user_score_sports = data_from_2013[(data_from_2013['genre'] == 'Sports') & (data_from_2013['user_score'].notna())]['user_score']

results = st.ttest_ind(
    user_score_action, 
    user_score_sports,
    equal_var=False)

print('p-value:', results.pvalue)

if results.pvalue < alpha:
    print("Failed to reject the null hypothesis")
else:
    print("Reject the null hypothesis")
```

**To test the hypotheses, a method was used to test the equality of the average of two general populations according to their samples. The first hypothesis says that the means of the two samples are equal, as a result the number p turned out to be very small, and accordingly the average values ​​of the samples are not included in 95% of the values.
