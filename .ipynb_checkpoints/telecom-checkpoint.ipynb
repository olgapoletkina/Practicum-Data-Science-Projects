## Project description

**Telecom Operator Client Churn Prediction Project**

**Project Overview**

A telecom operator seeks to predict customer churn. If there's a risk of a customer leaving, they will be offered promotional codes and special terms. The team has collected personal data on some clients, along with their tariff and contract information.

**Service Description**

**Main Services Offered:**

- **Landline Telephone Service:** Possibility of connecting a telephone device to multiple lines.
- **Internet Services:** Connection types include DSL (Digital Subscriber Line) and Fiber Optic.

**Additional Services:**

- **Internet Security:** Antivirus (DeviceProtection) and unsafe site blocking (OnlineSecurity).
- **Dedicated Technical Support Line (TechSupport).**
- **Online Backup:** Cloud storage for data backup.
- **Streaming Services:** Streaming TV (StreamingTV) and movie catalog (StreamingMovies).

Clients can pay monthly or sign contracts for 1–2 years. Various billing methods are available, including electronic checks.

**Data Description**

The data consists of files from different sources:
- `contract_new.csv` – Contract information.
- `personal_new.csv` – Personal data of the client.
- `internet_new.csv` – Information about Internet services.
- `phone_new.csv` – Information about telephony services.

In all files, the `customerID` column contains the client's code. The contract information is up to date as of February 1, 2020.

**Data Files Information:**

- **personal_new.csv - Personal Data of Client**
    - `gender` – Gender.
    - `SeniorCitizen` – Senior citizen status by age.
    - `Partner` – Presence of a spouse/partner.
    - `Dependents` – Presence of dependents.
- **contract_new.csv - Contract Information**
    - `BeginDate` – Date when the client started using services.
    - `EndDate` – Date when the client stopped using services.
    - `Type` – Contract type: monthly, yearly, etc.
    - `PaperlessBilling` – Billing via email.
    - `PaymentMethod` – Payment method.
    - `MonthlyCharges` – Monthly expenses on services.
    - `TotalCharges` – Total amount spent on services.
- **internet_new.csv – Internet Services Information**
    - `InternetService` – Availability of Internet service.
    - `OnlineSecurity` – Network firewall.
    - `OnlineBackup` – Online data backup storage.
    - `DeviceProtection` – Antivirus.
    - `TechSupport` – Dedicated technical support line.
    - `StreamingTV` – Online TV.
    - `StreamingMovies` – Online movie streaming.
- **phone_new.csv - Telephony Services Information**
    - `MultipleLines` – Ability to connect a telephone device to multiple lines simultaneously.

Create a target feature using the `contract` table. When the value of the `end_date` parameter is equal to `No`, the client contact is considered current and then the target attribute is `0`, otherwise `1`

```python
def churn(x):
    if x == 'No':
        return '0'
    else:
        return '1'

contract['churn'] = contract['end_date'].apply(churn)
```

```python
contract['end_date'] = contract['end_date'].replace('No', '2020-02-01')
```

```python
contract.end_date = pd.to_datetime(contract.end_date)
contract.begin_date = pd.to_datetime(contract.begin_date)
```

```python
contract['total_charges'] = pd.to_numeric(contract['total_charges'], errors='coerce')
```

## Dividing data to train and test.

```python
features = data.drop('churn', axis=1)
target = data['churn']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    features,
    target,
    test_size=0.25,
    random_state=RANDOM_STATE,
    stratify=target
)
print('X_train:', X_train.shape)
print('X_test:', X_test.shape)
print('y_train:', y_train.shape)
print('y_test:', y_test.shape)
```

### Feature importance analysis

```python
phik_overview = data.phik_matrix()
phik_overview['churn'].sort_values(ascending=False)
```

## Model selection.

### Categorical features.

```python
X_train.columns.to_series().groupby(X_train.dtypes).groups
```

```python
categorical_data = ['type', 'paperless_billing', 'payment_method', 'gender', 'senior_citizen', 'partner', 
                    'dependents', 'internet_service', 'online_security', 'online_backup', 'device_protection', 
                    'tech_support', 'streaming_tv', 'streaming_movies', 'multiple_lines']
```

```python
X_train.iteritems = X_train.items
X_test.iteritems = X_test.items
```

```python
for col in categorical_data:
    X_train[col] = X_train[col].astype('category')
    X_test[col] = X_test[col].astype('category')
```

### CatBoostClassifier

#### Parameters for CatBoostClassifier.

```python
parameters_cbc = {
    'max_depth': [4, 5, 6],
    'learning_rate': [0.01, 0.05, 0.1]
}

model_cbc = CatBoostClassifier(cat_features=categorical_data, class_weights=[0.3, 0.7], 
                               random_seed=RANDOM_STATE, verbose=0)

search_cbc = RandomizedSearchCV(model_cbc, parameters_cbc, cv=5, scoring='roc_auc', random_state=RANDOM_STATE)
search_cbc.fit(X_train, y_train)

best_params = search_cbc.best_params_
print('Best parameters:', best_params)
```

#### ROC AUC metric for CatBoostClassifier

```python
best_score_cbc = search_cbc.best_score_
print('Average ROC AUC score for CatBoostClassifier:', best_score_cbc)
```

### lightgbm.LGBMClassifier

#### Parameters for LGBMClassifier

```python
parameters_lgbmc = {
    'max_depth': range(5, 16),
    'num_leaves': [10, 20, 30, 40, 50],
    'learning_rate': [0.01, 0.05, 0.1],
    'n_estimators': [10, 50, 100, 150, 200],
    'min_child_samples': [10, 20, 30, 40, 50]
}

model_lgbmc = LGBMClassifier(random_seed=RANDOM_STATE, verbose=-1)

search_lgbmc = RandomizedSearchCV(model_lgbmc, parameters_lgbmc, cv=5, scoring='roc_auc', random_state=RANDOM_STATE)
search_lgbmc.fit(X_train, y_train)

best_params = search_lgbmc.best_params_
print('Best parameters:', best_params)
```

#### ROC AUC metric for LGBMClassifier

```python
best_score_lgbmc = search_lgbmc.best_score_
print('Average ROC AUC score for LGBMClassifier:', best_score_lgbmc)
```

### RandomForestClassifier

#### Apply OneHotEncoder for RandomForestClassifier

```python
ohe_features = X_train.select_dtypes(include='category').columns.to_list()

encoder_ohe = OneHotEncoder(drop='first', handle_unknown='error', sparse=False)

column_transformer = ColumnTransformer(
    transformers=[('encoder', encoder_ohe, ohe_features)],
    remainder='passthrough'
)

column_transformer.fit(X_train)
X_train_ohe = column_transformer.transform(X_train)

# pd.DataFrame(X_train_ohe).head(3)
```

#### Parameters for RandomForestClassifier

```python
warnings.filterwarnings("ignore")

parameters_rfc = {
    'max_depth': [5, 15],
    'n_estimators': [50, 100, 150, 200],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2'],
    'bootstrap': [True, False]
}

model_rfc = RandomForestClassifier(random_state=RANDOM_STATE, verbose=0)

search_rfc = RandomizedSearchCV(model_rfc, parameters_rfc, cv=5, scoring='roc_auc', n_iter=50, random_state=RANDOM_STATE)
search_rfc.fit(X_train_ohe, y_train)

best_params_rfc = search_rfc.best_params_
print('Best parameters:', best_params_rfc)
```

#### ROC AUC metric for RandomForestClassifier

```python
best_score_rfc = search_rfc.best_score_
print('Average ROC AUC score for RandomForestClassifier:', best_score_rfc)
```

### Checking the ROC-AUC metric of the best CatBoostClassifier model

```python
predict_proba_cbc = search_cbc.best_estimator_.predict_proba(X_test)[:, 1]
roc_auc_cbc = roc_auc_score(y_test, predict_proba_cbc)
print('ROC AUC:', roc_auc_cbc)
```

The metric `ROC AUC` on the test sample showed a result that satisfies the customer’s requirement

```python
fpr_cbc, tpr_cbc, thresholds_cbc = roc_curve(y_test.astype(int), predict_proba_cbc)

plt.figure(figsize=(10, 5))
plt.plot(fpr_cbc, tpr_cbc)
plt.plot([0, 1], [0, 1], linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive')
plt.ylabel('True Positive')
plt.title('ROC-AUC CatBoostClassifier')
plt.show()
```
